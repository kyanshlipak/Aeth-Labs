```python
%matplotlib ipympl 
#matplotlib interactive for IPython

#import libraries and functions
import pandas as pd
import math
import matplotlib.pyplot as plt # to make mpl faster
import matplotlib as mpl # to make mpl faster
import numpy as np #numpy for tons of stuff
import matplotlib.style as mplstyle
mpl.rcParams['path.simplify_threshold'] = 1.0 # to make mpl faster
mplstyle.use('fast') # to make mpl faster
import AlicatMFC as alicat #main python file to use functions from there
from scipy.signal import savgol_filter # savgol filter for smoothing


#customize plot font sizes
plt.rc('axes',titlesize=10)
plt.rc('xtick',labelsize=7)
plt.rc('ytick',labelsize=7)
plt.rc('legend',fontsize = 8)
plt.rc('figure',titlesize = 15)
plt.style.use('dark_background')

#create "mosaic" customizing grid layout. Semicolons separate rows
fig,axes = plt.subplot_mosaic("A;B")
fig.set_size_inches(14,17)
plt.tight_layout
plt.subplots_adjust(wspace =0.25,hspace = 0.55)
plt.ion()

def plotAll(variable,MAs,axis,smoothed):
    if smoothed:
        fig.suptitle(variable + " smoothed")
        axis.plot(savgol_filter(MAs[0][variable],20,3), label = "1 (350)", color = "red")
        axis.plot(savgol_filter(MAs[1][variable],20,3), label = "2 (350)", color = "orange")
        axis.plot(savgol_filter(MAs[2][variable],20,3), label = "3 (350)", color = "green")
        axis.plot(savgol_filter(MAs[3][variable],20,3), label = "4 (200)", color = "cyan")
        if len(MAs) == 5:
            axis.plot(savgol_filter(MAs[4][variable],20,3), label = "5 (200)", color = "violet")
    else:
        fig.suptitle(variable)
        axis.plot(MAs[0][variable], label = "1 (350)", color = "red")
        axis.plot(MAs[1][variable], label = "2 (350)", color = "orange")
        axis.plot(MAs[2][variable], label = "3 (350)", color = "green")
        axis.plot(MAs[3][variable], label = "4 (200)", color = "cyan")
        if len(MAs) == 5:
            axis.plot(MAs[4][variable], label = "5 (200)", color = "violet")
```
